```python
# Smart Parking Lot Management System

def parking_system(capacity, logs):
    parked = 0
    peak = 0
    for log in logs:
        if log == "IN":
            parked += 1
        elif log == "OUT" and parked > 0:
            parked -= 1
        if parked > peak:
            peak = parked
    print("Currently Parked Vehicles:", parked)
    if parked > capacity:
        print("Parking Status: Full")
    else:
        print("Parking Status: Available")
capacity = 50
logs = ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"]
parking_system(capacity, logs)
```

```python
# Online Food Delivery Time Estimator

def delivery_time(distance, traffic, weather):
    time = distance * 5
    if traffic == "High":
        time += 15
    elif traffic == "Medium":
        time += 10
    if weather == "Rainy":
        time += 10
    elif weather == "Storm":
        time += 20
    print("Estimated Delivery Time:", time, "minutes")
delivery_time(8, "High", "Rainy")
```

```python
# Movie Theatre Seat Occupancy Analyzer

def seat_occupancy(total_seats, booked_list):
    booked = len(booked_list)
    percentage = (booked / total_seats) * 100
    print("Occupancy:", percentage, "%")
    if percentage == 100:
        print("Show Status: Housefull")
    elif percentage >= 75:
        print("Show Status: Almost Full")
    else:
        print("Show Status: Available")
seat_occupancy(200, [1]*150)
```

```python
# Cloud Server Load Classification System

def server_load(cpu):
    avg = sum(cpu) / len(cpu)
    print("Average CPU Load:", avg, "%")
    if avg < 50:
        print("Server Status: Normal")
    elif avg <= 80:
        print("Server Status: Warning")
    else:
        print("Server Status: Critical")
server_load([45,60,70,85,90])
```

```python
# Smart Classroom Resource Usage Monitor

def resource_monitor(resources):
    overused = []
    for name, hours in resources.items():
        if hours > 8:
            overused.append(name)
    print("Overused Resources:", overused)
    if overused:
        print("Energy Alert: Yes")
    else:
        print("Energy Alert: No")
resources = {
"Projector":6,
"AC":9,
"Lights":4
}
resource_monitor(resources)
```

```python
# Online Event Registration Capacity Controller

def event_registration(capacity, registrations):
    if registrations > capacity:
        waitlist = registrations - capacity
        print("Confirmed Registrations:", capacity)
        print("Waitlisted Users:", waitlist)
        print("Registration Status: Closed")
    else:
        print("Confirmed Registrations:", registrations)
        print("Waitlisted Users: 0")
        print("Registration Status: Open")
event_registration(100,105)
```
