```python
# Employee Performance Bonus Eligibility

employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}
highest = max(employees.values())
top_performers = []
for name in employees:
    if employees[name] == highest:
        top_performers.append(name)
print("Top Performers Eligible for Bonus:", ", ".join(top_performers), "(Score:", highest, ")")
```

```python
# Search Query Keyword Analysis

query = "Buy mobile phone buy phone online"
query = query.lower()
words = query.split()
frequency = {}
for word in words:
    if word in frequency:
        frequency[word] += 1
    else:
        frequency[word] = 1
result = {}
for word in frequency:
    if frequency[word] > 1:
        result[word] = frequency[word]
print(result)
```

```python
# Sensor Data Validation

sensor_readings = [3, 4, 7, 8, 10, 12, 5]
valid = []
for i in range(len(sensor_readings)):
    if sensor_readings[i] % 2 == 0:
        valid.append((i, sensor_readings[i]))
print("Valid Sensor Readings (Hour,Value):", valid)
```

```python
# Email Domain Usage Analysis

emails = [
"ravi@gmail.com",
"anita@yahoo.com",
"kiran@gmail.com",
"suresh@gmail.com",
"meena@yahoo.com"
]
domain_count = {}
for email in emails:
    domain = email.split("@")[1]
    if domain in domain_count:
        domain_count[domain] += 1
    else:
        domain_count[domain] = 1
total = len(emails)
for domain in domain_count:
    percentage = (domain_count[domain] / total) * 100
    print(domain, ":", int(percentage), "%")
```

```python
# Sales Spike Detection

sales = [1200, 1500, 900, 2200, 1400, 3000]
average = sum(sales) / len(sales)
limit = average * 1.3
for i in range(len(sales)):
    if sales[i] > limit:
        print("Day", i+1, ":", sales[i])
```

```python
# Duplicate User ID Detection

user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]
count = {}
for user in user_ids:
    if user in count:
        count[user] += 1
    else:
        count[user] = 1
for user in count:
    if count[user] > 1:
        print(user, "→", count[user], "times")
```
