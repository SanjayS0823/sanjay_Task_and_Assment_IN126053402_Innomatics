```python
# Smart Light Controller

class SmartLight:
    def __init__(self, light_name):
        self.light_name = light_name
        self.status = "OFF"
    def turn_on(self):
        self.status = "ON"
    def turn_off(self):
        self.status = "OFF"
    def display_status(self):
        print(self.light_name, "is", self.status)

name = input("Enter Light Name: ")
action = input("Enter Action (ON/OFF): ")
light = SmartLight(name)
if action.upper() == "ON":
    light.turn_on()
else:
    light.turn_off()
light.display_status()
```

```python
# Employee ID Card System

class Employee:
    def __init__(self, name, emp_id, dept):
        self.name = name
        self.emp_id = emp_id
        self.dept = dept
    def display(self):
        print("\nEmployee ID Card")
        print("Name:", self.name)
        print("ID:", self.emp_id)
        print("Department:", self.dept)

name = input("Enter Employee Name: ")
emp_id = input("Enter Employee ID: ")
dept = input("Enter Department: ")
emp = Employee(name, emp_id, dept)
emp.display()
```

```python
# Mobile Contact Record

class Contact:
    def __init__(self, name, phone):
        self.name = name
        self.phone = phone
    def display(self):
        print("\nContact Saved")
        print("Name:", self.name)
        print("Phone:", self.phone)

name = input("Enter Contact Name: ")
phone = input("Enter Phone Number: ")
c = Contact(name, phone)
c.display()
```

```python
# Product Price Tag Generator

class Product:
    def __init__(self, name, price):
        self.name = name
        self.price = price
    def display(self):
        print("\nProduct:", self.name)
        print("Price: ₹", self.price)

name = input("Enter Product Name: ")
price = input("Enter Price: ")
p = Product(name, price)
p.display()
```

```python
# Movie Rating Display System

class Movie:
    def __init__(self, name, rating):
        self.name = name
        self.rating = rating
    def display(self):
        print("\nMovie:", self.name)
        print("Rating:", self.rating, "/ 5")

name = input("Enter Movie Name: ")
rating = input("Enter Rating: ")
m = Movie(name, rating)
m.display()
```
