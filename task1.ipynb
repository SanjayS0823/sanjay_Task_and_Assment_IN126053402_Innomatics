```python
# User Login Check

username = "admin"
password = "1234"

user_name = "admin"
user_password = "1234"

if user_name == username:
    if user_password == password:
        print("Login Successful")
    else:
        print("Invalid Credentials")
else:
    print("Invalid Credentials")
```

```python
# Pass / Fail Analyzer

marks = [45, 78, 90, 33, 60]
pass_count = 0
fail_count = 0
for m in marks:
    if m >= 50:
        pass_count = pass_count + 1
    else:
        fail_count = fail_count + 1
print("Passed Students:", pass_count)
print("Failed Students:", fail_count)
```

```python
# Simple Data Cleaner

names = [" Alice ", "bob", " CHARLIE "]
cleaned_names = []
for name in names:
    name = name.strip()
    name = name.lower()
    cleaned_names.append(name)
print("Cleaned Names:", cleaned_names)
```

```python
# Message Length Analyzer

messages = ["Hi", "Welcome to the platform", "OK"]
for msg in messages:
    print("Message:", msg)
    print("Length:", len(msg))
    if len(msg) > 10:
        print("Message is longer than 10 characters")
```

```python
# Error Message Detector

logs = ["INFO", "ERROR", "WARNING", "ERROR"]
error_count = 0
for log in logs:
    if log == "ERROR":
        error_count = error_count + 1
print("Total ERROR messages:", error_count)
```
