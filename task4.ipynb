```python
# Mobile Recharge Validation System

def recharge():
    valid_plans = [199, 299, 399, 599]
    while True:
        amount = int(input("Enter recharge amount: "))

        if amount < 50:
            print("Invalid: Minimum recharge is 50")
        elif amount in valid_plans:
            print("Recharge Successful")
            break
        else:
            print("Invalid Plan, Try Again")
recharge()
```

```python
# Inventory Reorder Alert System

def check_stock(products):
    for item in products:
        if products[item] < 15:
            print(item, ": Reorder Alert")
        else:
            print(item, ": Stock OK")
products = {
    "Rice": 10,
    "Oil": 25,
    "Salt": 12
}
check_stock(products)
```

```python
# Student Result Processing System

def result(marks):
    total = 0
    for n in marks:
        total = total + n
    average = total / len(marks)
    print("Average:", average)
    if average >= 50:
        print("Pass")
    else:
        print("Fail")
marks = [60, 70, 40]
result(marks)
```

```python
# Cab Fare Estimator

def fare():
    while True:
        km = float(input("Enter distance: "))
        peak = input("Peak hour? (yes/no): ")
        total = 50 + (12 * km)
        if peak == "yes":
            total = total + (total * 0.25)
        print("Total Fare:", total)
        retry = input("Retry? (yes/no): ")
        if retry == "no":
            break
fare()
```

```python
# Attendance Eligibility

def check_attendance(att):
    present = 0
    for day in att:
        if day == "P":
            present = present + 1
    percentage = (present / len(att)) * 100
    print("Attendance:", percentage)
    if percentage >= 75:
        print("Eligible")
    else:
        print("Not Eligible")
attendance = ["P", "A", "P", "A", "P"]
check_attendance(attendance)
```

```python
# Password Strength Checker

def check_password(password):
    special = "@#$"
    has_digit = False
    has_special = False
    for ch in password:
        if ch.isdigit():
            has_digit = True
        if ch in special:
            has_special = True
    if len(password) >= 8 and has_digit and has_special:
        print("Strong Password")
    else:
        print("Weak Password")
check_password("hello@098")
```
