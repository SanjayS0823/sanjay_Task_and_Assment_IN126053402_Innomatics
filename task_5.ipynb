```python
# Post Engagement Analyzer

def check_engagement(likes):
    total = 0
    for like in likes:
        total = total + like
    print("Total Likes:", total)
    if total >= 1000:
        print("Post Status: Viral Post")
    else:
        print("Post Status: Normal Engagement")
likes = [200, 300, 250, 300]
check_engagement(likes)
```

```python
# Medicine Stock Alert

def check_stock(stock):
    print("Medicine Stock:", stock)
    if stock < 10:
        print("Status: Low Stock Alert")
    else:
        print("Status: Stock Sufficient")
check_stock(6)
```

```python
# Rainfall Checker

def rainfall_check(data):
    total = 0
    for rain in data:
        total = total + rain
    average = total / len(data)
    print("Average Rainfall:", average)
    if average >= 70:
        print("Rainfall Status: Adequate Rainfall")
    else:
        print("Rainfall Status: Inadequate Rainfall")
rainfall_check([60, 80, 75])
```

```python
# Duplicate Account Detection

def check_duplicate(users):
    if len(users) != len(set(users)):
        print("Duplicate Accounts Found: Yes")
    else:
        print("Duplicate Accounts Found: No")
check_duplicate(["user1", "user2", "user1"])
```

```python
# Appointment Eligibility

def check_age(age):
    print("Patient Age:", age)
    if age >= 18:
        print("Eligibility Status: Eligible")
    else:
        print("Eligibility Status: Not Eligible")
check_age(21)
```

```python
# Premium Crop Filter

def premium_filter(prices):
    premium = []
    for price in prices:
        if price > 2000:
            premium.append(price)
    print("Premium Crops:", premium)
premium_filter([1500, 2500, 3200, 1800])
```

```python
# Application Health Checker

def check_health(errors):
    print("Error Count:", errors)
    if errors == 0:
        print("System Status: Healthy")
    elif errors <= 5:
        print("System Status: Minor Issues")
    else:
        print("System Status: Critical Issues")
check_health(7)
```

```python
# Transaction Limit Checker

def check_transaction(amount):
    print("Transaction Amount:", amount)
    if amount <= 50000:
        print("Transaction Status: Approved")
    else:
        print("Transaction Status: Rejected")
check_transaction(60000)
```

```python
# Attendance Eligibility

def check_attendance(att):
    present = 0
    for day in att:
        if day == "P":
            present = present + 1
    percentage = (present / len(att)) * 100
    print("Attendance Percentage:", percentage)
    if percentage >= 75:
        print("Exam Eligibility: Eligible")
    else:
        print("Exam Eligibility: Not Eligible")
check_attendance(["P", "P", "A", "P", "P"])
```

```python
# Electricity Bill Analyzer

def electricity_bill(units):
    bill = 0
    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + ((units - 100) * 5)
    else:
        bill = (100 * 3) + (100 * 5) + ((units - 200) * 7)
    print("Total Bill:", bill)
    if bill < 500:
        print("Usage: Low Usage")
    elif bill <= 1500:
        print("Usage: Moderate Usage")
    else:
        print("Usage: High Usage")
electricity_bill(250)
```
